7. Crie uma nova coluna chamada Total que seja o produto entre as colunas

8. Agrupe o DataFrame pela coluna Categoria e calcule o total de produtos
por categoria.

```python
import pandas as pd

#criando o data
data = {
    'Produto': ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E'],
    'Categoria': ['Eletrônicos', 'Eletrônicos', 'Roupas', 'Roupas', 'Eletrônicos'],
    'Preço': [10.00, 25.00, 5.00, 15.00, 30.00],
    'Quantidade': [100, 50, 200, 75, 25]
}

df = pd.DataFrame(data)

#agrupando o dataframe pela coluna categoria e calculando o total de produtos por categoria.
total_por_categoria = df.groupby('Categoria')['Quantidade'].sum()

display(total_por_categoria)

```

11. Preencha os valores nulos de uma coluna numérica com a mediana dessa
coluna.

```python
import pandas as pd

#preencher os valores nulos de uma coluna numérica com a mediana dessa coluna.
data = {
    'Coluna': [1, 2, None, 4, 5]
}

df = pd.DataFrame(data)

#encontrar a mediana da coluna
mediana = df['Coluna'].median()

#preencher os valores nulos com a mediana
df['Coluna'].fillna(mediana, inplace=True)

display(df)
```

13. Crie uma tabela dinâmica a partir do DataFrame, calculando a soma de
Quantidade por Categoria.

```python
import pandas as pd

#criando o dataframe
data = {
    'Produto': ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E', 'Produto F'],
    'Categoria': ['Eletrônicos', 'Eletrônicos', 'Roupas', 'Roupas', 'Eletrônicos', 'Roupas'],
    'Preço': [10.00, 25.00, 5.00, 15.00, 30.00, 20.00],
    'Quantidade': [100, 50, 200, 75, 25, 150]
}

df = pd.DataFrame(data)

#cria uma tabela dinâmica calculando a soma de Quantidade por Categoria
pivot_table_categoria = pd.pivot_table(df, values='Quantidade', index='Categoria', aggfunc='sum')

display(pivot_table_categoria)
```

14. Mescle dois DataFrames baseados em uma coluna comum chamada ID
utilizando um join.

```python
import pandas as pd

# Criando o primeiro DataFrame de exemplo
data1 = {'ID': [1, 2, 3, 4, 5],
         'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Eva']}
df1 = pd.DataFrame(data1)

# Criando o segundo DataFrame de exemplo
data2 = {'ID': [1, 2, 3, 6, 7],
         'Cidade': ['Nova Iorque', 'Londres', 'Paris', 'Tóquio', 'Sydney']}
df2 = pd.DataFrame(data2)

# Mesclando os DataFrames com base na coluna 'ID' usando um inner join
df_mesclado = pd.merge(df1, df2, on='ID', how='inner')

# Exibindo o DataFrame mesclado
display(df_mesclado)
```

15. Crie um DataFrame com dados faltantes e use a interpolação linear para
preenchê-los.

```python
import pandas as pd
import numpy as np


# Criando um DataFrame com dados faltantes usando np.nan
data = {'Coluna1': [1, 2, np.nan, 4, 5, np.nan, 7]}
df = pd.DataFrame(data)

print("DataFrame original com dados faltantes:")
display(df)

# Preenchendo os valores nulos usando interpolação linear
# A interpolação funciona porque np.nan em uma coluna de números inteiros resulta em um tipo float, que é numérico
df_interpolado = df.interpolate(method='linear')

print("\nDataFrame com dados faltantes preenchidos por interpolação linear:")
display(df_interpolado)
```

16. Crie um gráfico de barras a partir de um DataFrame usando a biblioteca
Pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Criando um DataFrame de exemplo
data = {'Categoria': ['A', 'B', 'C', 'D', 'E'],
        'Valor': [23, 45, 56, 12, 39]}
df = pd.DataFrame(data)

# Criando um gráfico de barras a partir do DataFrame
df.plot(x='Categoria', y='Valor', kind='bar')

# Adicionando título e rótulos aos eixos
plt.title('Exemplo de Gráfico de Barras')
plt.xlabel('Categoria')
plt.ylabel('Valor')

# Exibindo o gráfico
plt.show()
```

18. Exiba apenas as linhas de um DataFrame onde o nome do produto começa
com a letra ”B”.

```python
import pandas as pd

# Criando um DataFrame de exemplo
data = {'Produto': ['Produto A', 'Produto B', 'Produto C', 'Banana', 'Produto D'],
        'Preço': [10.00, 25.00, 5.00, 15.00, 30.00]}
df = pd.DataFrame(data)

print("DataFrame original:")
display(df)

# Filtrando as linhas onde o nome do produto começa com "B"
df_filtrado = df[df['Produto'].str.startswith('B')]

print("\nDataFrame filtrado (Produtos que começam com 'B'):")
display(df_filtrado)
```

19. Crie um DataFrame a partir de um dicionário de listas e calcule a média
de cada coluna numérica.

```python
import pandas as pd

# Criando um dicionário de listas
data = {'Coluna A': [10, 20, 30, 40, 50],
        'Coluna B': [5, 15, 25, 35, 45],
        'Coluna C': ['A', 'B', 'C', 'D', 'E']}

# Criando o DataFrame a partir do dicionário
df = pd.DataFrame(data)

print("DataFrame original:")
display(df)

# Calculando a média de cada coluna numérica
media_colunas = df.mean(numeric_only=True)

print("\nMédia de cada coluna numérica:")
display(media_colunas)
```

20. Agrupe os dados por uma coluna categórica e calcule a mediana de uma
coluna numérica.

```python
import pandas as pd

# Criando um DataFrame de exemplo
data = {'Categoria': ['A', 'B', 'A', 'C', 'B', 'C'],
        'Valor': [10, 15, 12, 18, 20, 25]}
df = pd.DataFrame(data)

print("DataFrame original:")
display(df)

# Agrupando por 'Categoria' e calculando a mediana de 'Valor'
mediana_por_categoria = df.groupby('Categoria')['Valor'].median()

print("\nMediana de 'Valor' por 'Categoria':")
display(mediana_por_categoria)
```
